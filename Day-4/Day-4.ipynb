###################################################### if-else statement #################################################
The if-else statement is a control structure that lets your program choose between two actions based on
whether a condition is true or false.

```python
age=18
if age>18:
    print("You are an adult.")
elif age==18:
    print("You are exactly 18 years old.")
else:
    print("You are a minor.")
# This code checks the age and prints a message based on the age category.
```

```python
age=int(input("Enter your age: "))
if age > 18:
    print("You are an adult.")
elif age == 18:
    print("You are exactly 18 years old.")
else:
    print("You are a minor.")
    
```

```python
age=int(input("Enter your age: "))
if age>18 and age<30:
    print("You are an adult.")
elif age>=30 and age<50:
    print("You are middle-aged.")
elif age>=50:
    print("You are a senior citizen.")
else:
    print("You are a minor.")
    
```

**while Loop**
With the while loop we can execute a set of statements as long as a condition is true.

```python
i=1
while i<10:
    print("Number is:", i)
    i += 1
```

```python
#Table of 5 using while loop
i=1
l=5

while i <= 10:
    print("Number is:",l,"X",i,"=" ,i*l)
    i += 1
```

```python
i=1
while i <= 100:
    if i % 2 == 0:
        print("Even number is:", i)
    else:
        print("Odd number is:", i)
    i += 1
```

```python
i=1
while i<=20:
    print("Number is:", i)
    j=1
    while j<=10:
        print( i, "X", j, "=", i*j)
        j += 1

    i += 1
```

```python
i=1
while i<=100:
    if i%2==0:
        print("Even number is:", i)
    else:
        print("Odd number is:", i)
    i += 1
```

```python
#break Statement
#With the break statement we can stop the loop even if the while condition is true:
i=1
while i<=20:
    if i==10:
        break
    print("Number is:", i)
    i += 1
```

```python
# continue Statement
# With the continue statement we can stop the current iteration, and continue with the next:
i=1
while i<=20:
    i += 1
    if i==10:
        continue
    print("Number is:", i)
    
```
